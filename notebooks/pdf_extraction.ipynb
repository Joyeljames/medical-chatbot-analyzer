```python
import fitz  # PyMuPDF

pdf_path = "../data/raw/sample_medical_report.pdf"
```

```python
pdf_document = fitz.open(pdf_path)
print("pages:", len(pdf_document))
```

```python
for page_num in range(len(pdf_document)):
    print(page_num)
```

```python
page  = pdf_document.load_page(0)
text = page.get_text()
print(text)
```

```python
len(text)
```

```python
full_text = ""
for page_num in range(len(pdf_document)):
    page  = pdf_document.load_page(page_num)
    text = page.get_text()
    full_text += text + "\n"

    pdf_document.close()
```

```python
print(full_text[:800])  # Print the first 1000 characters of the full text
```

```python
len(full_text)
```

```python
words  = full_text.split()
len(words)
```

```python
with open(
    "../artifacts/extracted_text/report.txt", "w",
    encoding="utf-8"
) as f:
    f.write(full_text)
```

```python
with open(
    "../artifacts/extracted_text/report.txt", "r",
    encoding="utf-8"
) as f:
    extracted_text = f.read()

print(extracted_text[:800])  # Print the first 1000 characters of the extracted text
```
