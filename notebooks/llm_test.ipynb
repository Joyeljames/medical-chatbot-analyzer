```python
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(
    model="llama3"
)
```

```python
response = llm.invoke(
    "what is cholestrol"
)
```

```python
print(response.content)
```

```python
response_new = llm.invoke("Explain machine learning in one paragraph.")
print(response_new.content)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)

db = FAISS.load_local(
    "../artifacts/Faiss",
    embedding_model,
    allow_dangerous_deserialization=True
)
```

```python
retriever = db.as_retriever(
    search_kwargs = {"k":3}
)
```

```python
question = "what is my cholestrol level"
```

```python
response = retriever.invoke(question)
print(response)
```

```python
context = "\n\n".join(
    [doc.page_content for doc in response]
)
```

```python
print(context)
```

```python
prompt = f"""
you are a medical report assistant

Answer only using context

Context: {context}

Questions: {question}
""" 
```

```python
response_llm = llm.invoke(
    prompt
) 

print(response_llm.content)
```
