```python
with open(
    "../artifacts/clean_text/clean_report.txt","r",
    encoding = "latin-1") as f:

    text = f.read()
```

```python
print(text[:500])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 50
)
```

```python
chunks = splitter.split_text(text)
```

```python
print("total chunks", len(chunks))
```

```python
for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}")
    print("-"*50)
    print(chunk)
```

```python
with open("../artifacts/chunks/chunks_report.txt","w",
              encoding="latin-1") as f:
    
    for i,chunk in enumerate(chunks):
        f.write(f"\n\n===== CHUNK {i+1} ======\n")
        f.write(chunk)
        
```
