```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddding = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS
```

```python
db = FAISS.load_local(
    "../artifacts/Faiss",
    embeddding,
    allow_dangerous_deserialization=True
)
```

```python
print(type(db))
```

```python
retreiver = db.as_retriever(
    search_kwargs = {"k":3}
)
```

```python
question = "what is my cholestrol level"
```

```python
docs = retreiver.invoke(question)
```

```python
for doc in docs:
    print(doc.page_content)
    print("-"*50)
```

```python
questions = [
    "what is my cholestrol level?",
    "what is vitamin situation?",
    "what are the fiondings"
]
```

```python
for q in questions:
    document = retreiver.invoke(q)
    print("Questions :",q)
    
    for doc in document:
        print(doc.page_content)
        print("-"*50)
```
