```python
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(
    model="llama3"
)
```

```python
with open(
    "../artifacts/clean_text/clean_report.txt",
    "r",
    encoding="latin-1"
) as f:
    
    report = f.read()
```

```python
prompt = f"""
Yoy are an expert medical assistant.

Summarize this report

report : {report}

provide:
1 . key findings
2 . abnormal values
3 . Recommendations
"""
```

```python
response = llm.invoke(prompt)

print(response.content)
```
